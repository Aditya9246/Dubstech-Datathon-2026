```python
import pandas as pd
from pathlib import Path

current_dir = Path.cwd()
#file_path = '/Users/nipun/Projects/Health_Datathon'
# 1. Reload the data
data_path = current_dir.parent / "data" / "NHIS_Data.csv"

print(f"Loading data from: {data_path}")
df = pd.read_csv(data_path, engine="pyarrow")

# 2. Check the types
print(df.dtypes)
```

```python
# Strip spaces from all column names
df.columns = df.columns.str.strip()
cols_to_drop = ['SUBTOPIC', 'SUBTOPIC_ID', 'NESTING_LABEL', 'NESTING_LABEL_ID', 'FLAG', 'ESTIMATE_TYPE_ID', 'TIME_PERIOD_ID', 'STANDARD_ERROR']
df.drop(columns=cols_to_drop, inplace=True,)
df.head()
```
